```python
URL =  []
url = driver.find_elements(By.XPATH,'//div[@class="property "]/a[1]')
for i in url:
    URL.append(i.get_attribute('href'))
```

```python
BHK_type = []
Property_type = []
Beds = []
Bathroom = []
Balcony = []
Address = []
City = []
State_code = []
Country = []
Landmark = []
Carpet_Area = []
Status = []
Floor = []
Transaction_Type = []
Year_built = []
Price_per_sqaure_feet = []
Furnished_status = []
Facing = []
Car_Parking = []
Type_of_ownership = []
Booking_amount = []
Buy_total_price = []
```

```python
for i in URL:
    driver.get(i)
    time.sleep(5)
    try:
        bhk_type = driver.find_element(By.XPATH,'//span[@class="parameter"][1]')
        BHK_type.append(bhk_type.text)
    except NoSuchElementException:
        BHK_type.append('-')

    try:
        p_type = driver.find_element(By.XPATH,'//li[@class="flexbox flex-align-center"][2]/div[2]/a')
        Property_type.append(p_type.text)
    except NoSuchElementException:
        Property_type.append('-')

    try:
        bed = driver.find_element(By.XPATH,'//div[@class="parameters"]/span[1]')
        Beds.append(bed.text)
    except NoSuchElementException:
        Beds.append('-')

    try:
        bath = driver.find_element(By.XPATH,'//li[@class="flexbox flex-align-center"][3]/div[2]')
        Bathroom.append(bath.text)
    except NoSuchElementException:
        Bathroom.append('-')

    try:
        add = driver.find_element(By.XPATH,'//div[@class="no-wrap small-long-address"]')
        Address.append(add.text)
    except NoSuchElementException:
        Address.append('-')

    try:
        city = driver.find_element(By.XPATH,'//span[@class="city"]')
        City.append(city.text.replace('','Chicago'))
    except NoSuchElementException:
        City.append('-')

    try:
        c_area = driver.find_element(By.XPATH,'//div[@class="parameters"]/span[3]')
        Carpet_Area.append(c_area.text)
    except NoSuchElementException:
        Carpet_Area.append('-')

    try:
        status = driver.find_element(By.XPATH,'//div[@class="status-label "]/div')
        Status.append(status.text)
    except NoSuchElementException:
        Status.append('-')

    try:
        year = driver.find_element(By.XPATH,'//li[@class="flexbox flex-align-center"][11]/div[2]')
        Year_built.append(year.text)
    except NoSuchElementException:
        Year_built.append('-')

    try:
        price_per_SF = driver.find_element(By.XPATH,'/html/body/div[1]/div/div[2]/div/div[2]/div[1]/div[2]/div[2]/div/div/div[1]/div[1]/div[5]/div[2]')
        Price_per_sqaure_feet.append(price_per_SF.text)
    except NoSuchElementException:
        Price_per_sqaure_feet.append('-')

    try:
        total_price = driver.find_element(By.XPATH,'//div[@class="no-wrap"]')
        price_text = total_price.text
        price_inr = float(price_text.replace('$','').replace(',', ''))
        Buy_total_price.append(price_inr * 83.12)
    except NoSuchElementException:
        Buy_total_price.append('-')  
```
